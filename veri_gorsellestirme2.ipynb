# Boxplot

### Veri seti hikayesi

total_bill : yemeğin toplam fiyatı

tip : bahşiş

sex : ücret ödeyen kişilerin cinsiyeti (0 = male, 1 = female)

smoker : gelen grupta sigara içen müşteri var mı? (0 = no, 1 = yes)

day : gün(3 = thur, 4 = friday, 5 = saturday, 6 = sunday)

time : nezaman (0 = day, 1 = dinner)

size : gruptaki kişi sayıları

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
sns.boxplot(x = df["total_bill"]);
```

Grafikte soldan sağa sırası ile ilk çizgi minimum değeri göstermektedir. Kutunun sol çizgisi %25'i, kutunun ortasındaki çizgi %50'yi, kutunun sağ çizgisi %75'i gösterir. en sağ taraftaki nokta nokta kısımlar bize aykırı durumları gösterir. En yüksek değeri aykırı değer de olsa yine de en sağdaki noktadır.

## Çizgi grafikleri

Çizgi grafikleri, grafik teknikleri arasında bilinmesi gereken ve görece diğerlerine göre daha zor veri görselleştirme türüdür. Zamana bağlı değişkenlerin olduğu veri setlerinde ve nesnelerin internetisenaryolarında makinelerin ürettiği verileri görselleştirmek için kullanılır.

FMRI beyne bağlanan bir cihaz aracılığıyla elde edilen bir veri setidir.

subject : verilerin toplandığı kişileri belli eder

timepoint : zaman noktalarını ifade eder.

evemt : verilerin toplanmasıyla ilgili olayları ifade eder.

region : verilerin toplandığı bölgeyi ifade ediyor.

sinyal : sinyali ifade eder.

```python
import seaborn as sns

fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()
```

```python
df.shape
```

```python
df["subject"].value_counts()
```

```python
df.dtypes
```

Görüldüğü üzere timepoint ve signal numeric değerlere sahiptir. Diğerleri object yani categorical değişkene aittir. Timepoint datası int type'dadır ve kesikli sayı değerlerine sahiptir. Bu yüzden incelenmesi gereken asıl şey signal değerleridir. Signal değerlerinin değişimine bakarak diğer değişkenlerin nasıl değiştiğini gözlemlememiz gerekmektedir.

```python
df.describe().T
```

```python
df.signal.describe()
```

```python
df.groupby("timepoint")["signal"].count()
```

```python
df.groupby("signal").count()
```

```python
df.groupby("timepoint")["signal"].describe()
```

```python
sns.lineplot(x = "timepoint",y ="signal", data = df);
```

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event", data = df);
```

İncelediğimiz veri seti mekanik bir settir. Iot gibi bir veriseti. Yukarıda timepoint ve signal incelendiğinde ve sonrasında timepoint ve signale categorik değişken event eklendiğinnde, eventte bulunan stim ve cue değişkeninin stim pik ve dip noktalarını oluşturduğunu gözlemlemekteyiz. 

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event", style = "event", data = df);
```

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event", style = "event", markers = True, dashes = False, data = df);
```

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "region", style = "event", data = df);
```
