## Examine correlations with the target

#### Our linear regression model is going to predict the median_house_value based on the other features. In order to be able to do this, there needs to be some kind of mathematical relationship - or correlation - between the features and the target.

#### A clever statistical calculation called the Pearson product moment correlation is usually used to calculate this. It returns a correlation coefficient. If this is close to 1 then theres a perfect correlation, if its close to 0 then theres little or no correlation, while if its less than 1 theres a negative correlation.

```python
import warnings
warnings.simplefilter("ignore")

plt.figure(figsize = (14,8))
corr = df.corr();
sns.heatmap(corr, annot=True, cmap = "YlGnBu")
plt.show()
```

## Create the regression model 
#### Define X and y

```python
X = df.drop(["median_house_value", "ocean_proximity"],axis = 1)
y = df["median_house_value"]
X.head()
```

```python
X_train, X_test,y_train, y_test = train_test_split(X,y,test_size = 0.3,random_state = 1)
```

## Scale the data
At the moment, the data in our dataframe consist of small and large values. Weve got some columns, such as proximity_near_bay which contain a 1 or 0, while others like total_rooms have median values of over 2650. As these vary so much in size, they could confuse our model. To put them all on an even scale, well use the StandardScaler() function. Importantly, we ise fit_transform() on our X_train data, but only use fit() on our X_test data. Not doing this can cause data leakage and may give away the answer to our model.

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Create the linear regression model

```python
model = LinearRegression()
model.fit(X_train,y_train)
LinearRegression()
```

## Generate predictions from your model

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

## Calculating the Root Mean Squared Error
The standard metric for assessing the performance of a regression model is called the Root Mean Squared Error or RMSE. 

```python
print("RMSE : ",np.sqrt(mean_squared_error(y_test,y_pred)))
```

## Plotting the predicted versus actual values

```python
test = pd.DataFrame({"Predicted_value" : y_pred, "Actual_value" : y_test})
fig = plt.figure(figsize=(16,8))
test = test.reset_index()
test = test.drop(["index"],axis = 1)
plt.plot((test[:50]))
plt.title("predicted vs actual values")
plt.legend(["Actual value","Predicted value"]);
```
