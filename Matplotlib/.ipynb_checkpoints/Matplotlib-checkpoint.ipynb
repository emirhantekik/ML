# Which one should you use? (pyplot vs matplotlib 00 method?)

* When plotting something quickly, okay to use the pyplot method
* When plotting something more advenced, use the OO method

```python
hearth_disease.head()
```

```python
over_50 = hearth_disease[hearth_disease["age"] > 50 ]
over_50.head()
```

```python
len(over_50)
```

```python
## pyplot method
over_50.plot(kind = "scatter",
            x = "age",
            y = "chol",
            c = "target");
```

```python
##OO method

fig, ax  = plt.subplots(figsize=(10, 6))
over_50.plot(kind = "scatter",
            x = "age",
            y = "chol",
            c = "target",
            ax = ax);
## ax.set_xlim([45, 100])
```

```python
## important
fig ,ax = plt.subplots(figsize = (10, 6))

##plot the data
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                    c = over_50["target"])

## Customize the plot
ax.set(title = "Hearth Disease and Cholesterol Levels",
      xlabel = "Age",
      ylabel = "Cholesterol")

## Add the legend
ax.legend(*scatter.legend_elements(),title = "Target");

## Add a horizontal line 
ax.axhline(over_50["chol"].mean(),
          linestyle = "--")
```
