# Grouping

In general, grouping data in Pandas works as follows:

df.groupby(by=grouping_columns)[columns_to_show].function()

```python
columns_to_show = ["Total day minutes","Total eve minutes","Total night minutes"]
df.groupby(["Churn"])[columns_to_show].describe(percentiles = [])
```

```python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]
df.groupby(["Churn"])[columns_to_show].agg([np.mean, np.std, np.min , np.max])
```

```python
pd.crosstab(df["Churn"],df["State"])
```

```python
pd.crosstab(df["Churn"],df["State"],normalize=True)
```

# This will resemble pivot tables to those familiar with Excel. And, of course, pivot tables are implemented in Pandas: 
### the pivot_table method takes the following parameters:

values â€“ a list of variables to calculate statistics for,

index â€“ a list of variables to group data by,

aggfunc â€“ what statistics we need to calculate for groups, e.g. sum, mean, maximum, minimum or something else.

```python
df.pivot_table(
    ["Total day calls", "Total eve calls", "Total night calls"],
    ["Area code"],
    aggfunc="mean",
)
```
